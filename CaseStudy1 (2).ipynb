```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency
```

```python
# Read the Excel file
df = pd.read_csv('D:\Any other\Revathy\Data_CS1.csv')
```

```python


# The company wishes to clarify whether there is any increase in sales after stepping into digital marketing.
# We can do this by comparing the mean sales before and after digital marketing.

sales_before = df['Sales_before_digital_add(in $)'].mean()
sales_after = df['Sales_After_digital_add(in $)'].mean()

print(f"Average sales before digital marketing: {sales_before}")
print(f"Average sales after digital marketing: {sales_after}")

# Visualize the sales before and after digital marketing
df_melt = df.melt(id_vars=['Month', 'Region', 'Manager'], value_vars=['Sales_before_digital_add(in $)', 'Sales_After_digital_add(in $)'], var_name='Marketing Type', value_name='Sales')
sns.boxplot(x='Marketing Type', y='Sales', data=df_melt)
plt.show()

# The company needs to check whether there is any dependency between the features “Region” and “Manager”.
# We can check this using a chi-square test of independence.

contingency_table = pd.crosstab(df['Region'], df['Manager'])
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")

# If the p-value is less than 0.05, we reject the null hypothesis and conclude that there is a significant relationship between Region and Manager.
# Otherwise, we fail to reject the null hypothesis and conclude that we don't have enough evidence to show a significant relationship.
```
