```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df = pd.read_excel('D:\\Any other\\Revathy\\Data_1.xls')
```

```python
# Display the columns in the dataset
print("Columns in the dataset:")
print(df.columns)
```

```python
# Calculate and print the mean of each numerical column
numerical_cols = df.select_dtypes(include=[np.number])
print("\nMean of each numerical column:")
print(numerical_cols.mean())
```

```python
# Print the total null values in each column
print("\nTotal null values in each column:")
print(df.isnull().sum())
```

```python
# 1. Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='SL', y='SW', hue='Classification', data=df)
plt.title('Scatter plot: Sepal Length vs Sepal Width')
plt.show()

# 2. Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Classification', y='SL', data=df)
plt.title('Boxplot for Sepal Length among different classes')
plt.show()

# 3. Pair plot
sns.pairplot(df, hue='Classification')
plt.show()
```
